## Нахождение минимума функции методом параболической интерполяции

```python
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

prec = 1E-7


#Зададим исходную функцию, имеющую несколько различных локальных минимумов (по ней удобно будет проверить программу)
def f(x):
    f = 3 * math.sin(4 * x) + 2 * math.cos(7 * x)
    return f

#Зададим функцию для нахождения минимума нашим методом
def minfun(f, a, b):
    while True:
        m = (a + b) / 2
        q1 = ((m - a) ** 2) * (f(m) - f(b)) - ((m - b) ** 2) * (f(m) - f(a)) # - считаем числитель и знаменатель отдельно, чтобы повысить точность вычислений
        q2 = 2 * ((m - a) * (f(m) - f(b)) - (m - b) * (f(m) - f(a)))
        q = q1 / q2
        if abs (q) <= prec:
            break
        elif q >= 0:
            b = m
        else:
            a = m
    return m
```

Проверим теперь программу на нескольких отрезках функции и построим график

```python
#Минимум функции на отрезке [1, 2]
print (minfun(f, 1, 2))    
```

```python
#Построим график с указанием минимума на заданном отрезке
xx = np.linspace(0, 3, 100)
pm = minfun(f, 1, 2)
plt.plot(xx, [f (xk) for xk in xx], '-')
plt.plot(pm, f (pm), 'o')
plt.title ('Минимум функции на отрезке')
plt.grid(True)
plt.show()
```
