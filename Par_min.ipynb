## Нахождение минимума функции методом параболической интерполяции

```python
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

prec = 1E-10
mid = (a + b) / 2

#Зададим исходную функцию
def f(x):
    f = 4 * math.sin(8 * x) + 5 * math.cos(4 * x)
    return f

#Зададим функцию для нахождения минимума нашим методом
def minfun(f, a, b, m, prec):
    q = m - (((m - a) ** 2) * (f(m) - f(b)) - ((m - b) ** 2) * (f(m) - f(a))) / (2 * ((m - a) * (f(m) - f(b)) - (m - b)*(f(m) - f(a))))
    if np.abs (q - m) < prec:
        return (q + m) / 2 
    elif f(q) < f(m) :
        return minfun(f, m, q, b, prec)
    else:
        return minfun(f, a, m, q, prec)
```

```python
#Минимум функции на заданном отрезке
print (minfun(f, -1, 1, mid, prec))
```

```python
#Построим график с указанием минимума на заданном отрезке
xx = np.linspace(-0.5, 2, 100)
pm = minfun(f, -1, 1, mid, 1E-10)
plt.plot(xx, [f (xk) for xk in xx], '-')
plt.plot(pm, f (pm), 'o')
plt.title ('Минимум функции на отрезке')
plt.grid(True)
plt.show()
```
