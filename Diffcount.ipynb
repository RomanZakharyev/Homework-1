```python
import numpy
import matplotlib
import matplotlib.pyplot as plt

'Введём исходную функцию, зададим какое-нибудь значение x0 и вычислим производные этой функции тремя способами'

x0 = 1.5

def function(x):
    return x ** 11

#Вычисление производных численно
def differential_side(x0, h):
    dif = (function(x0 + h) - function(x0)) / h 
    return dif

def differential_center(x0, h):
    dif = (function(x0 + h) - function(x0 - h)) / (2 * h) 
    return dif

#Вычисление производной аналитически
def differential_analyt(x0):
    return 11 * x0 ** 10

'Зададим значения по осям графиков центрального и одностороннего приближения и построим их'

h_x = [10 ** (k / 10) for k in range ( -200, 0 )]
differential_side_y = [abs(differential_side(x0, i) - differential_analyt(x0)) for i in h_x]
differential_center_y = [abs(differential_center(x0, i) - differential_analyt(x0)) for i in h_x]

plt.loglog(h_x, differential_side_y, color='blue', marker='o', linestyle='--', label = 'Одностороннее прибл.')
plt.loglog(h_x, differential_center_y, color='red', marker='o', linestyle='--', label = 'Центральное прибл.')

plt.title('Демонстрация различных методов вычисления производной')
plt.xlabel('h')
plt.ylabel('Модуль разности производных')
plt.grid(True)
plt.legend()
plt.show()

```

Вывод: во-первых, по графику видно, что при h < 10^-16 (то есть меньше величины машинного эпсилона) значения становятся равны константе, так как там модуль разности производных будет порядка машинного эпсилона, во-вторых, можно заметить возникнувшие флуктуации в интервале от машинного эпсилона до минимума, связанные с тем, что при малых значениях h происходит неточность в округлении и, как следствие, накопление ошибки, а после минимума флуктуации пропадают в связи с нивелированием ошибок в округлении при больших h. Сравнивая два этих метода, нужно отметить, что центральное приближение даёт более точный результат, чем одностороннее, логичной причиной чего является меньшее значение минимума у центрального приближения. Также можно заметить, что угол наклона красного графика приблизительно в 2 раза больше синего, что демонстрирует зависимость центрального приближения от о малого порядка h^2, а одностороннего - порядка h, что и требовалось доказать.
