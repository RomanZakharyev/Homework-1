```python
import numpy
import matplotlib
import matplotlib.pyplot as plt

'Введём исходную функцию, зададим какое-нибудь значение x0 и вычислим производные этой функции тремя способами'

x0 = 2.5

def function(x):
    return x ** 11

#Вычисление производных численно
def differential_side(x, h):
    dif = (function(x + h) - function(x)) / h 
    return dif

def differential_center(x, h):
    dif = (function(x + h) - function(x - h)) / (2 * h) 
    return dif

#Вычисление производной аналитически
def differential_analyt(x):
    return 11 * x ** 10

'Зададим значения по осям графиков центрального и одностороннего приближения и построим их'

h_x = [10 ** (k / 10) for k in range ( -200, 0 )]
differential_side_y = [abs(differential_side(x0, i) - differential_analyt(x0)) for i in h_x]
differential_center_y = [abs(differential_center(x0, i) - differential_analyt(x0)) for i in h_x]

plt.loglog(h_x, differential_center_y, color='red', marker='o',  linestyle='-', label = 'Центральное прибл.')
plt.loglog(h_x, differential_side_y, color='blue', marker='o',  linestyle='-', label = 'Одностороннее прибл.')

plt.title('Демонстрация различных методов вычисления производной')
plt.xlabel('h')
plt.ylabel('Модуль разности производных')
plt.grid(True)
plt.legend()
plt.show()
```
